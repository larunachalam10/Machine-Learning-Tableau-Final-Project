```python
import pandas as pd
import datetime as dt
```

```python
data2016 = pd.read_csv("reservations2016.csv", encoding= 'unicode_escape')
data2016
```

```python
data2016.columns
```

```python
data2016["EntityType"].unique()
```

```python
data2016["UseType"].unique()
```

```python
data2016_clean = data2016.loc[(data2016["UseType"] == "Overnight") & (data2016["EntityType"] == "Site")]
```

```python
data2016_clean["SiteType"].unique()
```

```python
data2016_clean.columns
```

```python
data2016_clean = data2016_clean.drop(columns=['HistoricalReservationID', 
       'OrderNumber', 'OrgID', 'CodeHierarchy', 'RegionCode', 'ParentLocationID',
       'LegacyFacilityID', 'UseType', 'ProductID', 'EntityType', 'EntityID',
       'FacilityLongitude', 'FacilityLatitude', 'Tax', 'UseFee', 'TranFee',
       'AttrFee', 'TotalPaid', 'Tent', 'Popup', 'Trailer',
       'RVMotorhome', 'Boat', 'HorseTrailer', 'Car', 'FifthWheel', 'Van',
       'CanoeKayak', 'BoatTrailer', 'Motorcycle', 'Truck', 'Bus', 'Bicycle',
       'Snowmobile', 'OffRoadlAllTerrainVehicle', 'PowerBoat', 'PickupCamper',
       'LargeTentOver9x12', 'SmallTent', 'Marinaboat'])
```

```python
data2016_clean.dtypes
```

```python
data2016_clean["StartDate"] = pd.to_datetime(data2016_clean["StartDate"])
data2016_clean["EndDate"] = pd.to_datetime(data2016_clean["EndDate"])
data2016_clean["OrderDate"] = pd.to_datetime(data2016_clean["OrderDate"])
```

```python
data2016_clean.dtypes
```

```python
data2016_clean = data2016_clean.loc[(data2016_clean["StartDate"] >= "2016-05-01") 
                                    & (data2016_clean["EndDate"] <= "2016-09-30")]
data2016_clean 
```

```python
data2016_clean.to_csv("reservations_clean2016.csv", index = False, header=True)
```
